```python
import pandas as pd
import time
from selenium import webdriver
from time import sleep
import datetime
from selenium.webdriver.common.by import By
from datetime import date, timedelta
```

```python
# Create driver object. Opens browser window.
chromedriver_path = 'C:/Users/Tevet/chromedriver.exe'
driver = webdriver.Chrome(chromedriver_path)
```

```python
##Dynamic search for city,num of adults,children and rooms
city = "New York"
numOfAdults = 2
numOfChild = 0
numOfRooms = 0
currency = "USD"
```

```python
##In this cell we define functions that each one of them scrape something else

def get_hotelName(PATH):
    name = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            name.append(curHotel.find_element(By.XPATH,'.//*[@class="fcab3ed991 a23c043802"]').text)
        except:
            name.append(None)
    return name
            
def get_price(PATH):
    price = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            price.append(curHotel.find_element(By.XPATH,'.//*[@class="fcab3ed991 fbd1d3018c e729ed5ab6"]').text)
        except:
            price.append(None)
    return price
            
def get_review(PATH):
    review = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            review.append(curHotel.find_element(By.XPATH,'.//*[@class="d8eab2cf7f c90c0a70d3 db63693c62"]').text)
        except:
            review.append(None)
    return review

def get_rating(PATH):
    rating = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            rating.append(curHotel.find_element(By.XPATH,'.//*[@class="b5cd09854e f0d4d6a2f5 e46e88563a"]').text)
        except:
            rating.append(None)
    return rating


def get_score(PATH):
    score = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            score.append(curHotel.find_element(By.XPATH,'.//*[@class="b5cd09854e d10a6220b4"]').text)
        except:
            score.append(None)
    return score
    
def get_centerDist(PATH):
    distance = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            distance.append(curHotel.find_element(By.CSS_SELECTOR,"span[data-testid='distance']").text)
        except:
            distance.append(None)
    return distance
    
def get_numNight(PATH):
    night = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            night.append(curHotel.find_element(By.XPATH,'.//*[@class="d8eab2cf7f c90c0a70d3"]').text[0])
        except:
            night.append(None)
    return night

def get_stars(PATH):
    stars = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            stars.append(curHotel.find_element(By.XPATH,'.//*[@class="e4755bbd60"]').get_attribute("aria-label")[0])
        except:
            stars.append(None)
    return stars

def get_room(PATH):
    room = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            room.append(curHotel.find_element(By.XPATH,'.//*[@class="df597226dd"]').text)
        except:
            room.append(None)
    return room

def get_bed(PATH):
    bed = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            bed.append(curHotel.find_element(By.XPATH,'.//*[@class="cb5b4b68a4"]').text)
        except:
            bed.append(None)
    return bed

def get_breakfast(PATH):
    breakfast = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            breakfast.append(curHotel.find_element(By.XPATH,'.//*[@class="e05969d63d"]').text)
        except:
            breakfast.append("0")
    return breakfast

def get_sustainable(PATH):
    sustainable = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            sustainable.append(curHotel.find_element(By.XPATH,'.//*[@class="d8eab2cf7f be09c104ad"]').text[-1])
        except:
            sustainable.append("0")
    return sustainable

def get_cancellation(PATH):
    cancellation = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            cancellation.append(curHotel.find_element(By.XPATH,'.//*[@class="d506630cf3"]').text)
        except:
            cancellation.append("0")
    return cancellation

def get_tax(PATH):
    tax = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            tax.append(curHotel.find_element(By.CSS_SELECTOR,"div[data-testid='taxes-and-charges']").text[3:])
        except:
            tax.append(None)
    return tax

def get_price_before(PATH):
    price = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            price.append(curHotel.find_element(By.XPATH,'.//*[@class="c5888af24f e729ed5ab6"]').text)
        except:
            price.append('No Discount')
    return price

def get_location(PATH):
    location = []
    for curHotel in driver.find_elements(By.XPATH,PATH):
        try:
            location.append(curHotel.find_element(By.XPATH,'.//*[@class="f4bd0794db b4273d69aa"]').text)
        except:
            location.append(None)
    return location

def get_weekend(start_date,end_date):
    date_range = pd.date_range(start_date, end_date - pd.Timedelta(days=1), freq='D')
    weekends = (date_range.weekday >= 5).sum()
    return weekends
    
```

```python
def next_page(PATH):
    nextPage = PATH
    driver.find_element_by_xpath(nextPage).click()
```

```python
import datetime

current_date = datetime.datetime.now() 
lengthOfStay = 5 ##LOS
timeToTravel = 30 ##TTT
hotelsEachDate = 100
appended_data = [] 

for ttt in range(1, timeToTravel):
    # Iterate over LOS
    for los in range(1, lengthOfStay+1):
        checkin_date = current_date + datetime.timedelta(days=ttt)
        checkin_year = checkin_date.year
        checkin_month = checkin_date.month
        checkin_monthday = checkin_date.day
        checkout_date = checkin_date + datetime.timedelta(days=los)
        checkout_year = checkout_date.year
        checkout_month = checkout_date.month
        checkout_monthday = checkout_date.day
        days_until_checkin = (checkin_date - current_date).days
        url = f"https://www.booking.com/searchresults.html?ss={city}&checkin_year={checkin_year}&checkin_month={checkin_month}&checkin_monthday={checkin_monthday}&checkout_year={checkout_year}&checkout_month={checkout_month}&checkout_monthday={checkout_monthday}&group_adults={numOfAdults}&group_children={numOfChild}&no_rooms={numOfRooms}&selected_currency={currency}"
        driver.get(url)
        driver.maximize_window()
        sleep(1)
        
        counter = 0
        while counter < hotelsEachDate:
            name = get_hotelName('.//*[@class="d20f4628d0"]')
            price = get_price('.//*[@class="d20f4628d0"]')
            review = get_review('.//*[@class="d20f4628d0"]')
            rating = get_rating('.//*[@class="d20f4628d0"]')
            score = get_score('.//*[@class="d20f4628d0"]')
            distance = get_centerDist('.//*[@class="d20f4628d0"]')
            night = get_numNight('.//*[@class="d20f4628d0"]')
            star = get_stars('.//*[@class="d20f4628d0"]')
            room = get_room('.//*[@class="d20f4628d0"]')
            bed = get_bed('.//*[@class="d20f4628d0"]')
            breakfast = get_breakfast('.//*[@class="d20f4628d0"]')
            sustainable = get_sustainable('.//*[@class="d20f4628d0"]')
            cancellation = get_cancellation('.//*[@class="d20f4628d0"]')
            tax = get_tax('.//*[@class="d20f4628d0"]')
            priceBefore = get_price_before('.//*[@class="d20f4628d0"]')
            location = get_location('.//*[@class="d20f4628d0"]')
            weekend = get_weekend(checkin_date , checkout_date)
            
            counter += len(name)
            
            next_page('//*[@id="search_results_table"]/div[2]/div/div/div[4]/div[2]/nav/div/div[3]/button/span/span')
            sleep(3)
            
            appended_data.append(pd.DataFrame({
                    'Name': name,
                    'SnapShotDate': (f"{current_date.year}/{current_date.month}/{current_date.day}"),
                    'Checkin' : (f"{checkin_year}/{checkin_month}/{checkin_monthday}"),
                    'Checkout' : (f"{checkout_year}/{checkout_month}/{checkout_monthday}"),
                    'NumOfWeekendDays' : weekend,
                    'Nights':night,
                    'RoomType':room,
                    'BedType':bed,
                    'Breakfast':breakfast,
                    'Review':review,
                    'Rating':rating,
                    'Score':score,
                    'Cancellation':cancellation,
                    'Sustainable':sustainable,
                    'CenterDistance':distance,
                    'Location':location,
                    'Stars':star,
                    'Taxes':tax,
                    'PriceBdiscount':priceBefore,
                    'Price':price
                                       }))
        
df = pd.concat(appended_data)
df.to_csv(f'{city} Booking.com {current_date.day}.csv',index=False)
print(f"Succesfully saved {city} hotels as {city} Booking.csv ")
        
driver.quit()
```
