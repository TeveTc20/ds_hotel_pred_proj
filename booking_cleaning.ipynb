```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings("ignore")
```

```python
booking_df = pd.read_csv('Booking.csv')
```

```python
booking_df.head()
```

```python
booking_df.shape
```

```python
booking_df.info()
```

```python
booking_df['WebIndex'] = (booking_df.index % 100) + 1
```

```python
booking_df.isna().sum()
```

```python
##Dropping null values (For price and name we cant work without and others are not much nulls and we can allow to drop them)
booking_df = booking_df.dropna(subset=['Price'])
booking_df = booking_df.dropna(subset=['Name'])
booking_df = booking_df.dropna(subset=['Stars'])
booking_df = booking_df.dropna(subset=['BedType'])
booking_df = booking_df.dropna(subset=['Review'])
```

```python
##OneHotEncoding for Location column
booking_df['Location'] = booking_df['Location'].str.replace(', New York', '')
booking_df = pd.get_dummies(booking_df, columns = ['Location'])
```

```python
booking_df.columns
```

```python
# Changing the dates columns to datetime format
booking_df['SnapShotDate'] = pd.to_datetime(booking_df['SnapShotDate'],errors='coerce', format='%Y-%m-%d')
booking_df['Checkin'] = pd.to_datetime(booking_df['Checkin'],errors='coerce', format='%Y-%m-%d')
booking_df['Checkout'] = pd.to_datetime(booking_df['Checkout'],errors='coerce', format='%Y-%m-%d')
```

```python
##Checking Breakfast column values
booking_df['Breakfast'].unique()
```

```python
##Breakfast include = 1 , else = 0
booking_df['Breakfast'] = booking_df['Breakfast'].str.replace('Breakfast included','1')
booking_df['Breakfast'] = booking_df['Breakfast'].str.replace('Breakfast','1')
booking_df['Breakfast'] = booking_df['Breakfast'].fillna('0')
```

```python
##Checking Cancellation column values
booking_df['Cancellation'].unique()
```

```python
##Free cancellation include = 1 , else = 0
booking_df['Cancellation'] = booking_df['Cancellation'].str.replace('Free cancellation','1')
booking_df['Cancellation'] = booking_df['Cancellation'].str.replace('FREE cancellation • No prepayment needed','1')
booking_df['Cancellation'] = booking_df['Cancellation'].str.replace('Cancellation','1')
booking_df['Cancellation'] = booking_df['Cancellation'].fillna('0')
```

```python
##Dropping Sustainable column because every value is equal to 0 
booking_df.drop('Sustainable', inplace=True, axis=1)
```

```python
##Making Review column to show only the number of reviews
booking_df['Review'] = booking_df['Review'].str.replace('reviews','')
booking_df['Review'] = booking_df['Review'].str.replace(',', '')
```

```python
##Cleaning CenterDistance column to show only numbers(Meters parameters converted to km)
booking_df['CenterDistance'] = booking_df['CenterDistance'].str.replace('km from center','')
booking_df['CenterDistance'] = booking_df['CenterDistance'].str.replace('m from center','')
booking_df['CenterDistance'] = booking_df['CenterDistance'].replace('CenterDistance', np.nan)
```

```python
booking_df = booking_df.dropna(subset=['CenterDistance'])
```

```python
##Converting CenterDistance,Night and Score columns to numerical for fututre math operations
booking_df['CenterDistance'] = booking_df['CenterDistance'].apply(pd.to_numeric)
booking_df['Score'] = booking_df['Score'].apply(pd.to_numeric)
booking_df['Nights'] = booking_df['Nights'].apply(pd.to_numeric)
```

```python
##Converting the meters to km
booking_df['CenterDistance'] = booking_df['CenterDistance'].apply(lambda x: x/1000 if x > 100 else x)
```

```python
##Checking Rating column values
booking_df['Rating'].unique()
```

```python
##Function that find rating using score params
def score_to_rating(score):
    if score >= 9.5 and score <= 10:
        return 'Exceptional'
    elif score >= 9.0 and score < 9.5:
        return 'Excellent'
    elif score >= 8.5 and score < 9.0:
        return 'Wonderful'
    elif score >= 8.0 and score < 8.5:
        return 'Very Good'
    elif score >= 7.5 and score < 8.0:
        return 'Good'
    else:
        return 'Review score'
```

```python
booking_df['Rating'] = booking_df.apply(lambda row: score_to_rating(row['Score']) if pd.isna(row['Rating']) else row['Rating'], axis=1)
```

```python
##Converting Rating column to numerical
booking_df.replace({'Rating':{'Review score':5.0,'Good':6.0,'Very Good':7.0,'Excellent':8.0,
                        'Wonderful':9.0 ,'Exceptional':10.0}},inplace=True)
```

```python
##Filling Nights NaN values
booking_df['Nights'] = booking_df['Nights'].fillna((booking_df['Checkout'] - booking_df['Checkin']).dt.days)
```

```python
booking_df['Score'] = booking_df['Score'].replace('Score', np.nan)
```

```python
##Function that find score using rating params
def map_rating_to_score(rating):
    if rating == 'Exceptional': 
        return 9.5
    elif rating == 'Excellent':
        return 9.0
    elif rating == 'Wonderful':
        return 8.5
    elif rating == 'Very Good':
        return 8.0
    elif rating == 'Good':
        return 7.5
    else:
        return 6.0
```

```python
booking_df['Score'] = booking_df.apply(lambda row: map_rating_to_score(row['Rating']) if pd.isna(row['Score']) else row['Score'], axis=1)
```

```python
##Creating new columns of months and days separately for each date
booking_df['SnapShotMonth'] = booking_df['SnapShotDate'].dt.month.astype(str)
booking_df['SnapShotDay'] = booking_df['SnapShotDate'].dt.day.astype(str)
booking_df['CheckinMonth'] = booking_df['Checkin'].dt.month.astype(str)
booking_df['CheckinDay'] = booking_df['Checkin'].dt.day.astype(str)
booking_df['CheckoutMonth'] = booking_df['Checkout'].dt.month.astype(str)
booking_df['CheckoutDay'] = booking_df['Checkout'].dt.day.astype(str)
```

```python
booking_df['TTT'] = (booking_df['Checkin'] - booking_df['SnapShotDate']).dt.days
booking_df['LOS'] = booking_df['Nights'] + 1
```

```python
##Dropping the full date
booking_df.drop('SnapShotDate', inplace=True, axis=1)
booking_df.drop('Checkin', inplace=True, axis=1)
booking_df.drop('Checkout', inplace=True, axis=1)
```

```python
##Changing RoomType column to numerical 
def map_room_type_to_score(room_type):
    if 'Suite' in room_type:
        return 7
    elif 'King' in room_type:
        return 6
    elif 'Queen' in room_type:
        return 5
    elif 'Superior' in room_type:
        return 4
    elif 'Deluxe' in room_type:
        return 4
    elif 'Apartment' in room_type:
        return 4
    elif 'Quadruplet' in room_type:
        return 4
    elif 'Quadruple' in room_type:
        return 4
    elif 'Double' in room_type:
        return 3
    elif 'Empire' in room_type:
        return 3
    elif 'Guest' in room_type:
        return 3
    elif 'Twin' in room_type:
        return 3
    elif 'Standard' in room_type:
        return 3
    elif 'Studio' in room_type:
        return 3
    elif 'Classic' in room_type:
        return 3
    elif 'Comfort' in room_type:
        return 3
    elif 'Full' in room_type:
        return 3
    elif 'Triple' in room_type:
        return 3
    elif 'Accessible' in room_type:
        return 2
    elif 'Petite' in room_type:
        return 2
    elif 'Run' in room_type:
        return 2
    elif 'Interior' in room_type:
        return 2
    elif 'Single' in room_type:
        return 2
    elif 'Selected' in room_type:
        return 2
    elif 'Bunk' in room_type:
        return 1
    elif 'Murphy' in room_type:
        return 1
    else:
        return None
    
```

```python
booking_df['RoomType'] = booking_df['RoomType'].apply(lambda x: map_room_type_to_score(x))
```

```python
booking_df = booking_df.dropna(subset=['RoomType'])
```

```python
##Changing BedType column to numerical 
def bed_type_to_score(bed_type):
    if 'king' in bed_type:
        return 6
    elif 'queen' in bed_type:
        return 5
    elif 'full' in bed_type:
        return 4
    elif 'twin' in bed_type:
        return 4
    elif 'double' in bed_type:
        return 4
    elif 'Multiple' in bed_type:
        return 3
    elif 'bunk' in bed_type:
        return 2
    
    else:
        return bed_type
    
```

```python
booking_df['BedType'] = booking_df['BedType'].apply(lambda x: bed_type_to_score(x))
```

```python
##Cleaning Price column
booking_df['Price'] = booking_df['Price'].apply(lambda x : x.split('$')[1])
booking_df['Price'] = booking_df['Price'].str.replace(',', '')
```

```python
##Cleaning Taxes column
booking_df['Taxes'] = booking_df['Taxes'].apply(lambda x : x.split('$')[1])
booking_df['Taxes'] = booking_df['Taxes'].str.replace('taxes and charges', '')
```

```python
##Dropping PriceBdiscount because we wont use this params
booking_df.drop('PriceBdiscount',axis =1 ,inplace = True)
```

```python
booking_df.info()
```

```python
##Converting columns to numerical
booking_df['NumOfWeekendDays'] = booking_df['NumOfWeekendDays'].apply(pd.to_numeric)
booking_df['Breakfast'] = booking_df['Breakfast'].apply(pd.to_numeric)
booking_df['Review'] = booking_df['Review'].apply(pd.to_numeric)
booking_df['Cancellation'] = booking_df['Cancellation'].apply(pd.to_numeric)
booking_df['Stars'] = booking_df['Stars'].apply(pd.to_numeric)
booking_df['Taxes'] = booking_df['Taxes'].apply(pd.to_numeric)
booking_df['Price'] = booking_df['Price'].apply(pd.to_numeric)
booking_df['SnapShotMonth'] = booking_df['SnapShotMonth'].apply(pd.to_numeric)
booking_df['SnapShotDay'] = booking_df['SnapShotDay'].apply(pd.to_numeric)
booking_df['CheckinMonth'] = booking_df['CheckinMonth'].apply(pd.to_numeric)
booking_df['CheckinDay'] = booking_df['CheckinDay'].apply(pd.to_numeric)
booking_df['CheckoutMonth'] = booking_df['CheckoutMonth'].apply(pd.to_numeric)
booking_df['CheckoutDay'] = booking_df['CheckoutDay'].apply(pd.to_numeric)
```

```python
##Creating new column that check price per night
booking_df['PricePerNight'] = (booking_df['Price'] / booking_df['Nights']).round()
```

```python
booking_df['PricePerNight'] = booking_df['PricePerNight'].apply(pd.to_numeric)
```

```python
booking_df.drop('Location_Location',axis = 1,inplace = True)
```
